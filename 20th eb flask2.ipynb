# Q1. Explain GET and POST methods.

GET and POST are HTTP methods used to send data from a client (e.g. a web browser) to a server. The main difference between GET and POST is how the data is sent.

GET method:

Sends data in the URL query string.
Data is visible in the URL.
Limited amount of data can be sent.
Used to retrieve data from a server.
Can be cached by the browser.


POST method:

Sends data in the request body.
Data is not visible in the URL.
Can send large amounts of data.
Used to submit data to a server.
Cannot be cached by the browser.

In general, GET requests should be used for retrieving data and POST requests should be used for submitting data. However, there are some exceptions to this rule, such as when sending data to update a resource, in which case a POST request should be used even though it is not retrieving data.

# Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client to a different URL. When a client accesses a Flask route that calls redirect(), the server sends a response to the client with a status code of 302 (Found) and a Location header indicating the new URL to which the client should be redirected.

The redirect() function is often used in web applications to redirect the user to a different page after a form submission, a login/logout action, or any other action that requires a page change.

# Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are files that contain a combination of HTML, CSS, and other template languages (such as Jinja) that define the structure and content of a web page. In Flask, templates are typically used to separate the presentation layer from the application logic, allowing developers to create dynamic web pages without having to manually generate HTML.

The render_template() function in Flask is used to render a template and generate an HTML response that can be sent to the client. The function takes the name of a template file as its first argument, followed by any arguments to be passed to the template.
